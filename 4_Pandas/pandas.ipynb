### Series
* homogeneous
* vectorized operation
* handle missing value with NaN
* mutable value, immutable size

```python
#vectorized
s1 = pd.Series([1,2,3,4,5])
s2 = pd.Series([6,7,8,9,10])

print(s1 + s2)
```

```python
#immutable size, series not change the size instead it create another series
s1 = pd.Series([1,2,3,4,5])
s2 = pd.Series([6,7,8,9,10])

changed_s2 = s1.drop(0)
print(s1)
print(changed_s2)
```

### Query
rules
* expression as a string
* column based reference
* backticks for col_name with space or special chars
* operators ( &, |, ~, >=, >, <=, >, == etc)
* @ to reference python var
* chained comparisons

```python
#query -> copy
air_df.query("aqi > 100 & temperature > 30")[["city", "aqi"]]
```

```python
aqi = 100
air_df.query("aqi > @aqi & temperature > 30")[["city", "aqi"]]
```
