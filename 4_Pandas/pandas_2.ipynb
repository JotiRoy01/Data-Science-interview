### Apply
* it is a higher order function.
* it take another function as a input. 
* it provide the input one by one to another function

```python
#apply
df2 = df.copy()

df2["tax"] = df2["income"].apply(lambda x : "20%" if x >= 50000 else "10%")
```

```python
df2
```

### Map
* whe we want to map existing value to new value

```python
#map
gender_map = {"Male": "M", "Female": "F", "Unknown" : "U"}

df2["gender"] = df2["gender"].map(gender_map)
df2
```

```python
#Assign()
#-> to create a new column

df2 = df2.assign(new_column = df2["income"] * 1.1)
df2
```

```python
#replace
df2["country"] = df2["country"].replace("USA", "US")
```

```python
#Rename
df2.columns = ["Id", "Name", "Age", "Country", "Gender", "Income", "Tax", "New_Column"]
```

### Groupby & Aggregation

```python
df3.groupby("country")["income"].mean()
df3.groupby("country")["income"].min()
df3.groupby("country")["income"].max()

df3.groupby("gender")["income"].mean()
```

### agg() function
it's help to group multiple column and perform multiple operation

```python
df3.groupby("country")["income"].agg(["mean", "min", "max"])
```

```python
df3.groupby("country")["income"].agg(avg_salary = "mean", min_salary =  "min", max_salary = "max")
```

```python
df3.groupby("country").agg(
    {
        "income": "mean",
        "age": "mean"
    }
)
```

```python
df3.groupby("country").agg(
    max_salary = ("income", "max"),
    avg_age = ("age", "mean")
)
```

### Melt & Pivot

```python
new_df = pd.DataFrame({
    "country": ["US", "US", "BD", "IND"],
    "year": [2020, 2021, 2020, 2021],
    "sales": [100,120,90,110],
    "profit": [20, 25, 18, 22],
    "tax": [10, 15, 12, 10.6]
})
```

```python
new_df
```

```python
melt_df = new_df.melt(
     id_vars = ["country", "year"],
     value_vars = ["sales", "profit", "tax"],
     var_name = "matrics",
     value_name = "value"
)
```

```python
melt_df
```

```python
new_df = pd.DataFrame({
    "country": ["US", "US", "BD", "IND"],
    "year": [2020, 2021, 2020, 2021],
    "sales": [100,120,90,110],
    "profit": [20, 25, 18, 22]
})
```

### Merging & Joining

```python
df_customer = pd.DataFrame(
    {
        "customer_id": [1,2,3,4],
        "name": ["adam", "bob", "chrlie", "dave"]
    }
)

df_orders = pd.DataFrame({
    "order_id": [101, 102, 103, 104],
    "customer_id": [2,1,4,5],
    "amount": [250, 120, 300, 180]
})
```

```python
df_customer
```

```python
df_orders
```

```python
#inner join
pd.merge(df_customer, df_orders, on="customer_id")
```

```python
#left join
pd.merge(df_customer, df_orders, on="customer_id", how="left")
```

```python
#right join
pd.merge(df_customer, df_orders, on="customer_id", how="right")
```

```python
#outer join
pd.merge(df_customer, df_orders, on="customer_id", how="outer")
```

### Data concatenation
Stacking dataframes on top of each other  (rows) or side by side (columns)


```python
# concatentation
df1 = pd.DataFrame(
    {
        "id": [1,2,3],
        "name": ["adam", "eve", "bob"]
    }
)

df2 = pd.DataFrame(
    {
        "id": [4,5,6],
        "name": ["chrlie", "fester", "figian"]
    }
)
```

```python
df1
```

```python
df2
```

```python
#row wise
pd.concat([df1, df2]) 
```

```python
#create new index
pd.concat([df1, df2], ignore_index=True) 
```

```python
#columns wise
pd.concat([df1, df2], axis=1) 
```
