```python
#beutifulSoup
from bs4 import BeautifulSoup

with open("scraped_data1.html", "r") as f :
    html = f.read()
soup = BeautifulSoup(html, "lxml")
```

```python
soup.find("h1")
```

```python
all_h3 = soup.find_all("h3")
type(all_h3)
```

```python
all_countries = []

for h3 in all_h3 :
    name = h3.get_text(strip = True)
    population = h3.find_text("div").select("span.country-population")[0].get_text(strip = True)
    all_countries.append([name, population])
```

```python
import pandas as pd
df = pd.Dataframe(all_countries, columns = ["name", "population"])
```

```python
df.to_csv("cleaned_data/data.csv", index= False)
```

```python
page_count = 1
while True :
    URL = f"https://quotes.toscrape.com/page/{page_count}/"
    res = requests.get(URL)

    soup = BeautifulSoup(res.text)
    quotes = soup.select("div.quote")

    if not quotes :
        print("no valid pages anymore....")
        break

    with open(f"scraped_data/quotes{page_count}.html", "w") as f :
        f.write(res.text)
        print(f"download data from page {page_count}")
    
    page_count = page_count + 1
```

```python
#Extract information
with open("scraped_data/quotes1.html", "r") as f :
    html_content = f.read()
    all_quotes = soup.select("div.qoute")
    
```

```python
all_quotes = soup.select("div.quotes")
life_quotes = []

for q in all_quotes :
    all_tags = []

    for tag in q.select(".tags .tag") :
        all_tags.append(tag.get_text)

    if "life" in all_tags :
        text = q.select_one("span.text").get_text()
        author = q.select_one("small.author").get_text()
        life_quotes.append([text, author])

print(life_quotes)
```
