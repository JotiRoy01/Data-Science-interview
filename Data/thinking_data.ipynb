```python
import json
```

```python
#load the data
def load_data(filename) :
    with open(filename, "r") as f :
        data = json.load(f)
    return data
```

```python
py_data = load_data("store_data.json")
```

```python
#remove inconsistancy
def rm_incstcy(data) :
    txt_to_digit = {"one": 1, "two": 2, "three": 3, "four": 4, "five": 5}
    clean_data = []
    deduplication = set()
    for user in data :
        raw_rating = str(user["rating"]).strip().lower()
        if raw_rating in txt_to_digit :
            raw_rating = txt_to_digit[raw_rating]
        user["rating"] = raw_rating

        #handing missing value
        raw_age = user.get("age")
        if(raw_age == None) :
            user["age"] = raw_age

        #deduplication
        if(user["name"] in deduplication) :
            continue
        deduplication.add(user["name"])
        clean_data.append(user)
    return clean_data
```

```python
clean_data = rm_incstcy(py_data)
```

```python
#Get meaningful insights from data
def get_insights(data) :
    total_rating = 0
    for user in data  :
        total_rating += float(user["rating"])
    print(f"avg_rating {total_rating/len(data)}")

    #percentage of user with poor rating
    poor_rating = 0
    for user in data :
        if(float(user["rating"]) < 3) :
            poor_rating += 1
    print(f"% of user with poor rating = {poor_rating/len(data) * 100}%")
```

```python
get_insights(py_data)
```

```python
#Build a product recommendation feature(Rule based, not ML)
def get_recommendation(data) :
    recommendation = []
    for user in data :
        cur_recomdion = {}
        cur_recomdion["name"] = user["name"]
        if(float(user["rating"]) >= 4) :
            cur_recomdion["brand"] = "Apple"
        else :
            cur_recomdion["brand"] = "Samsung"
    recommendation.append(cur_recomdion)
    return recommendation
```

```python
get_recommendation(clean_data)
```
