```python
class Laptop :
    storage_type = "ssd"
    
    def __init__(self, RAM, storage) :
        self.RAM = RAM
        self.storage = storage

    @classmethod
    def get_storage_type(cls) :
        print(f"storage type = {cls.storage_type}")

    def get_info(self) : #instance method
        print(f"laptop has {self.RAM} RAM & {self.storage} {self.storage_type}")

l1 = Laptop("16gb", "512gb")
l2 = Laptop("8gb", "256gb")

l1.get_storage_type()
```

```python
class Laptop :
    storage_type = "ssd"
    
    def __init__(self, RAM, storage) :
        self.RAM = RAM
        self.storage = storage

    @classmethod
    def get_storage_type(cls) :
        print(f"storage type = {cls.storage_type}")

    def get_info(self) : #instance method
        print(f"laptop has {self.RAM} RAM & {self.storage} {self.storage_type}")

    @staticmethod
    def calc_discount(price, discount) :
        final_price = price - (discount * price / 100)
        print(f"final price = {final_price}")

l1 = Laptop("16gb", "512gb")
l2 = Laptop("8gb", "256gb")

l1.calc_discount(40_000, 10)
```

##### Product store  
design & create an online store for Products ( name, price)  
Track total products being created  
Create a static method to calculate discount on each product based on a % parameter

```python
class Product :
    total_products = 0
    
    def __init__(self, name, price) :
        self.name = name
        self.price = price
        Product.total_products += 1
    
    @staticmethod
    def calc_discount(price, discount_percent) :
        final_price = price - (discount_percent * price / 100)
        return final_price

# Create some products to test
p1 = Product("Laptop", 50000)
p2 = Product("Phone", 30000)
p3 = Product("Tablet", 20000)

print(f"Total products created: {Product.total_products}")
print(f"Price of {p1.name} after 15% discount: {Product.calc_discount(p1.price, 15)}")
```

### Encapsulation  
* wrappring data & functions into single unit
* data hiding  

```python
class BankAccount :
    def __init__(self, name, balance, password) :
        self.name = name #public
        self._balance = balance #protected
        self.__password = password # private - data mangling

        def get_balance(self) : #getter
            return self.__password

        def set_balance(self, new_Balance) :  # setter
            self.__password = 123
        

acc1 = BankAccount("Joti", 10_000)
print(acc1.name, acc1.balance)
print(acc1.name, acc1._BankAccount__balance)
```

### Inheritance
Reusing attr & methods from a parent (Base) class

```python
#single inheritance
class Employee :
    strat_time = "10am"
    end_time = "6pm"

    def change_time(self, new_end_time) :
        self.end_time = new_end_time

class Teacher(Employee) :
    def __init__(self, subject) :
        self.subject = subject

class AdminStaff(Employee) :
    def __init__(self, role) :
        self.role = role

t1 = Teacher("Math")

```

```python
#Multi level inheritance
class Employee :
    strat_time = "10am"
    end_time = "6pm"

    def change_time(self, new_end_time) :
        self.end_time = new_end_time

class AdminStaff(Employee) :
    def __init__(self, subject) :
        self.subject = subject

class Accountant(AdminStaff) :
    def __init__(self, salary, role) :
        super().__init__(role)
        self.salary = salary

acc1 = Accountant(25_000, "CA")
```

```python
#MUlticlass inheritance
class Teacher :
    def __init__(self, salary) :
        self.salary = salary

class Student :
    def __init__(self, gpa) :
        self.gpa = gpa

class TA(Teacher, Student) :
    def __init__(self, salary, gpa, name) :
        super().__init__(salary)
        Student.__init__(self, gpa)
        self.name = name


ta1 = TA(15_000, 9.3, "joti")

print(ta1.name, ta1.gpa, ta1.salary)
```

### Abstraction
Hiding internal details & showing only essential features

```python
from abc import ABC, abstractmethod

class Animal(ABC) :
    @abstractmethod
    def make_sound(self) :
        pass

class Lion(Animal) :
    def make_sound(self) :
        print("Roar....")

class Cow(Animal) :
    def make_sound(self) :
        print("Moo...")

lion = Lion()
lion.make_sound()

cow = Cow()
cow.make_sound()
```

```python
#function overriding
class Employee :
    def get_designation(self) :
        print("designation = Employee")

class Teacher(Employee):
    def get_designation(self) :
        print("designation = Teacher")

t1 = Teacher()
t1.get_designation()
```

```python
#Duck Typing
class Teacher() :
    def get_designation(self) :
        print("designation = Teacher")

class Accountant() :
    def get_designation(self) :
        print("designation = Accountant")

t1 = Teacher()
t1.get_designation()

acc1 = Accountant()
acc1.get_designation()
```
