```python
#Q4
def count_digit(n) :
    count = 0
    while(n > 0) :
        n = n // 10
        count += 1

    print(count)
```

```python
count_digit(321)
```

```python
#Q5
def print_digit(n) :
    while (n > 0) :
        num = n % 10
        num += num
        n = n//10
    return num
```

```python
print_digit(321)
```
