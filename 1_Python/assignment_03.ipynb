```python
#Q1
string = input("Enter a string")
count = 0
for i in range(len(string)) :
    if string[i] == string[len(string) - i - 1] :
        print(f"forward = {string[i]} and backword {string[len(string) - i - 1]}")
        count +=1
    if(count == len(string)) :
        print("Ther input string is a palindrome")
```

```python
#Q4
tup = (1,2,3,4,5,6,7,8,9)

even = []
odd = []

for i in range(len(tup)) :
    if(i % 2 == 0) :
        even.append(tup[i])
    else :
        odd.append(tup[i])

even = tuple(even)
odd = tuple(odd)
print(f"Even indexed elements: {even}")
print(f"Odd indexed elements: {odd}")
```

```python
#Q9
def print_duplicates(lst):
    seen = set()
    dupes = set()
    for x in lst:
        if x in seen:
            dupes.add(x)
        else:
            seen.add(x)
    if dupes:
        print(dupes)
    else:
        print("No duplicates")

# Example usage
print_duplicates([1, 2, 3, 2, 4, 1, 5])
```
