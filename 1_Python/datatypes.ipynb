#### Student Enrolments
Given a list of tuples with infor(name, subject): 
* list all unique course
* list students enrolled in english
* create dictionary (student, set of courese)

```python
info = [
    ("Alice", "Match"),
    ("Bob", "Science"),
    ("Alice", "Science"),
    ("Charlie", "Match"),
    ("Bob", "Match"),
    ("Alice", "English"),
    ("Charlie", "English")
]
```

```python
unique = set()
for tup in info :
    #print(tup[0])
    #print(tup[1])
    unique.add(tup[1])
```

```python
print(unique)
```

```python
for  name, course in info :
    if(course == "English"):
        print(name)
```

```python
dict1 = {}

for name, course in info :
    if(dict1.get(name) == None) :
        dict1.update({name: set()})
        dict1[name].add(course)
    else :
        dict1[name].add(course)
```
