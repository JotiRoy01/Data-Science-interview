### Match Case

```python
color = input("enter color: ")

match color :
    case "Green":
        print("Go")
    case "Yellow" :
        print("look")
    case "Red" :
        print("Stop")
    case _ : #default case defined by underscroe(_)
        print("wrong color")
```

### Break

```python
# break -> terminate the loop

i = 1
while( i <= 10) :
    if( i % 6 == 0) :
        break
        print(i)
        i += 1
print("outside loop now......")
```

### Continue

```python
#skip the current iteration

i = 1

while(i <= 10) :
    if(i % 3 == 0) :
        i += 1
        continue
    print(i)
    i += 1

print("outside the loop")
```

```python
#odd nums

i = 0

while(i <= 10) :
    i += 1
    if(i % 2 == 0) :
        continue
    print(i)
```

### For Loops

```python
# Used for sequential traversal
string = "hello"
#in -> membership operator
for var in string :
    print(var)

```

```python
word = "artificial intelligence"
ans = 0
for ch in word :
    if(ch == 'i') :
        ans += 1
print("count of i = ", ans)
```

### Lambda Function

```python
sum = lambda a, b : a + b # sum variable act as a function name, we can call it

print(sum(4,5))
```

```python

def cal_fact(n) :
    fact = 1
    for i in range(1, n + 1) :
        fact = fact * i
    return fact

n = int(input("enter the n "))
print(cal_fact(n))
```
