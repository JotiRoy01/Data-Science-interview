```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset("tips")
print(type(tips))
print(tips.head())
```

```python
sns.set_theme()
sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="smoker",
    style="smoker"
)
```

```python
#scatter plot
sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="time"
)
```

```python
#line plot
sns.lineplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="time"
)
```

```python
#line plot
fights = sns.load_dataset("flights")

sns.lineplot(
    data=fights,
    x="year",
    y="passengers"
)
```

```python
#bar plot
sns.barplot(
    data=tips,
    x="day",
    y="tip",
    hue="sex"
)
```

```python
#box plot
sns.boxenplot(
    data=tips,
    x="day",
    y="tip"
)
```

```python
#hist plot
penguins = sns.load_dataset("penguins")

sns.histplot(
    data=penguins,
    x = "body_mass_g",
    bins = 30
)
```

```python
#heatmap
pivot_flights = fights.pivot(index="month", columns="year", values="passengers")
pivot_flights
```

```python
sns.heatmap(
    pivot_flights,
    cmap="coolwarm",
    annot=True,
    fmt="d"
)

plt.title("Passengers Heatmap")
plt.tight_layout()
```

```python
fig, ax = plt.subplots()

sns.lineplot(
    data = tips,
    x= "day",
    y="total_bill",
    hue="sex",
    markers="o",
    errorbar=None,
    ax=ax
)

ax.set_title("avg bill value for each day")
ax.set_xlabel("days")
ax.set_ylabel("total bill amount")
fig.tight_layout()
```
